```python
import PyPDF2
from pprint import pformat
from pathlib import Path
from datetime import date
from typing import List

def get_pdf_files(directory:Path) -> List[Path]:
    """Return all pdf files in 'directory'."""

    directory = Path( r"C:\Users\Miranda and Warren\Documents\Invoice")
    if not directory.is_dir():
        raise TypeError()
    
    pdf_files = list(directory.glob('*.pdf'))
    return pdf_files

def create_unique_destination_path(
        directory:Path, 
        file_name: str = "extracted_data",
        suffix:str = '.txt',
        date: date = date.today()
    ) -> Path:
    """Return a file path with a unique YYMMDD date string suffix."""

    date_identifier = date.strftime("%y%m%d")
    unique_file_name = f"{file_name}_{date_identifier}{suffix}"
    unique_destination_path = directory / unique_file_name 
    return unique_destination_path

def extract_all_text(pdf_files: List[Path]) -> str:
    """Return all text from the pdf files in the list."""

    extracted_text = ""
    for pdf in pdf_files:
        with pdf.open(mode='rb') as pdf:
            pdf_reader = PyPDF2.PdfReader(pdf)
            page_object = pdf_reader.pages[0]
            page_text = page_object.extract_text()
            extracted_text += page_text
    
    return extracted_text
    
def save_data_as_python_module(
    destination:Path, 
    data:str,
    variable_name:str = 'data') -> None:
    
    destination.write_text(f"{variable_name} = {pformat(data,indent=2)}")

DIRECTORY = Path(r"C:\Users\Miranda and Warren\Documents\Invoice")

pdf_files = get_pdf_files(DIRECTORY)
extracted_text = extract_all_text(pdf_files)

destination_path = create_unique_destination_path(
    directory=DIRECTORY, 
    suffix='.py')

save_data_as_python_module(
    destination=destination_path,
    data=extracted_text)

```
